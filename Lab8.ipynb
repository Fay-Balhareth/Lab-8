```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Apply Edge Detection Operators: Use the Sobel, Roberts, and Prewitt operators to detect edges in the image.


```python
image = cv2.imread('/content/x-ray-skull-from-right-side.jpg', cv2.IMREAD_GRAYSCALE)

# Sobel Edge Detection
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)

plt.figure(figsize=(15, 8))
plt.subplot(1, 4, 1), plt.imshow(image, cmap='gray'), plt.title('original')
plt.subplot(1, 4, 2), plt.imshow(sobelx, cmap='gray'), plt.title('sobel X')
plt.subplot(1, 4, 3), plt.imshow(sobely, cmap='gray'), plt.title('sobel Y')
plt.subplot(1, 4, 4), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Sobel Edge Detection')
plt.show()
```

```python
# Roberts operator kernels
roberts_cross_x = np.array([[1, 0],
                            [0, -1]])
roberts_cross_y = np.array([[0, 1],
                            [-1, 0]])

# Apply the Roberts operator kernels
edge_x = cv2.filter2D(image, -1, roberts_cross_x)
edge_y = cv2.filter2D(image, -1, roberts_cross_y)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(edge_x**2 + edge_y**2)

plt.figure(figsize=(15, 8))
plt.subplot(1, 4, 1), plt.imshow(image, cmap='gray'), plt.title('original')
plt.subplot(1, 4, 2), plt.imshow(edge_x, cmap='gray'), plt.title('Roberts - Edge X')
plt.subplot(1, 4, 3), plt.imshow(edge_y, cmap='gray'), plt.title('Roberts - Edge Y')
plt.subplot(1, 4, 4), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Roberts Edge Detection')
plt.show()
```

```python
# Prewitt Operator
prewitt_x = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])
prewitt_y = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

# Apply Prewitt Operator
edge_x = cv2.filter2D(image, -1, prewitt_x)
edge_y = cv2.filter2D(image, -1, prewitt_y)

# Calculate the gradient magnitude
gradient_magnitude = np.sqrt(edge_x**2 + edge_y**2)

plt.figure(figsize=(15, 8))
plt.subplot(1, 4, 1), plt.imshow(image, cmap='gray'), plt.title('original')
plt.subplot(1, 4, 2), plt.imshow(edge_x, cmap='gray'), plt.title('Prewitt - Edge X')
plt.subplot(1, 4, 3), plt.imshow(edge_y, cmap='gray'), plt.title('Prewitt - Edge Y')
plt.subplot(1, 4, 4), plt.imshow(gradient_magnitude, cmap='gray'), plt.title('Prewitt Edge Detection')
plt.show()
```
